# Linear Regression

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

## Summary

* Regression is all about prediction.
* We can calculate the strength of an association (the *correlation*) between two variables.
* "Fundamental Theorem of Statistics": **Prediction = Reality + Error**

Often times, things will increase or decrease at a constant rate as other things increase or decrease too.

Other times, things increase or decrease at a continually-changing rate. 

In this section, we will be examining the first case.

When we examine how two variables, such as *x* and *y*, interact with each other, we are looking at the **correlation** of those variables. 

For instance, we might be interested in examining how much time spent on a smartphone correlates to performance in school.

## The Linear Correlation Coefficient, r

The <span style="color:blue">**linear correlation coefficient**</span>, *r*, is a numerical value with -1 &leq; r &leq; 1 that measures the **strength of linear correlation** of a 2-variable data set.

The closer *r* is to 1 (or to -1), the stronger the linear correlation. 

* If *r* is close to 1, that indicates a strong positive linear correlation. The data points seem to be "in a line" with a positive slope (*Note*: *r* is not the slope).
* If *r* is close to -1, that indicates a strong negative linear correlation. The data points seem to be "in a line" with a negative slope.
* If *r* is close to 0, that indicates no linear correlation.

For any given value of *r*, there is no universal way of classifying strength of linear correlation. 

We will use technology to determine the value of *r*. 

For instance, suppose we want to determine how strong of a correlation the following scatterplot shows

|x|y|
|:---:|:---:|
|1|1|
|2|3|
|3|4|
|4|6|

```python
x = [1,2,3,4]
y = [1,3,4,6]

fig, ax = plt.subplots()

ax.scatter(x, y, color='black')
ax.axvline(np.mean(x), color='red', linestyle='--')
ax.axhline(np.mean(y), color='blue', linestyle='--')
ax.grid(linestyle='--')
plt.show()
```

We can see (from the *Scatterplots* notes) that there seems to be a positive correlation. And while the data points look like they could all be very close in distance to a drawn line, just how close to that line would they be?

If you were wondering, below is the formula for determining *r*:

$$
r = \frac{\sum (x-\overline{x})(y-\overline{y})}{\sqrt{\sum(x-\overline{x})^2 \cdot \sum(y-\overline{y})^2}}
$$

As scary as that looks:

* The numerator is the total area of all rectangles formed by each point and the "new origin." (Note: The two points in the "new Quadrant II" would have negative areas.)
* The denominator is the square root of the product of squared deviations each data point is from the "new origin."

## Least-Squares Regression Equation

So now we know how to determine the strength of a linear relationship. The next step is to be able to make predictions using our data values. 

The <span style="color:blue">**least-squares regression equation**</span> (a.k.a. the *line of best fit*) is the equation of the line that **minimizes** the total squared distance each point is from the line.

In the scatterplot given before Example 1, the least squares regression equation would be 

$$ \hat{y} = 1.6x - 0.5$$

where $\hat{y}$ is our **predicted** value of *y* given an input value *x*.

Notice that the line goes through the "new origin."

```python
x = [1,2,3,4]
y = [1,3,4,6]

slope, intercept, r, pvalue, stderror = stats.linregress(x,y)

def leastsq(x):
    return slope*x + intercept

fig, ax = plt.subplots()

xvals = np.linspace(np.min(x), np.max(x), 200)  # x-coordinates to graph least-squares line

ax.scatter(x, y, color='black')
ax.plot(xvals, leastsq(xvals), color='purple')
ax.axvline(np.mean(x), color='red', linestyle='--')
ax.axhline(np.mean(y), color='blue', linestyle='--')
ax.grid(linestyle='--')
plt.show()

print(f"Regression Equation: y = {slope:0.4f}x + {intercept:0.4f}")
```

### Making Predictions with the Least-Squares Regression Equation

We can use the LSRE to make predictions about data values that were not observed (as well as ones that were).

For our dataset with 4 points (the one before Example 1), we found the equation to be $\hat{y} = 1.6x - 0.5$.

## Residuals

A <span style = "color:blue">**residual**</span> is the difference between the observed value and the expected/predicted value. 

$$\text{Residual} = \text{observed value} - \text{predicted value}$$
